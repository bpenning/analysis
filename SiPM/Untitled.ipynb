```python
import csv
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly
```

```python
def exponenial_func(x, a, b, c):
    return a*np.exp(-b*x)+c
```

```python
f2 = open('temp_vs_darkCurrent.txt', 'r') 
lines = f2.readlines()
x = []
y = [] 
for line in lines:
    p = line.split()
    y.append(float(p[0]))
    x.append(float(p[1]))
f2.close() 
print(x, y)
```

```python
my_dpi=200
plt.figure(figsize=(8, 6), dpi=100)
x_new = np.linspace(x[0], x[-1], num=len(x)*10) #creae new grid for x values to evaluate the fitting function ffit
coefs = poly.polyfit(x, y, 2) #perform fit
plt.plot(x_new, poly.polyval(x_new, coefs)) #plot fit, polyval gives y-values of the new x grid
plt.scatter(x, y, linewidth=2.0) #plot the data
plt.xlabel('Temp [K]')
plt.ylabel('Dark Current [nA]')
plt.show()
```
